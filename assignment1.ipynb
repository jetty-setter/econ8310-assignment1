```python
# assignment1.py
# Econ 8310 - Assignment 1
# Forecasting NYC Taxi Trips (Hourly)

import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# -----------------------------
# Load training and test data
# -----------------------------
train_url = "https://github.com/dustywhite7/econ8310-assignment1/raw/main/assignment_data_train.csv"
test_url = "https://github.com/dustywhite7/econ8310-assignment1/raw/main/assignment_data_test.csv"

train = pd.read_csv(train_url)
test = pd.read_csv(test_url)

# -----------------------------
# Prepare the data
# -----------------------------
# Use trips as dependent variable
y_train = train["trips"]

# -----------------------------
# Define the model
# -----------------------------
# Hourly data with weekly seasonality
# 24 hours * 7 days = 168
model = ExponentialSmoothing(
    y_train,
    trend="add",
    seasonal="add",
    seasonal_periods=168
)

# -----------------------------
# Fit the model
# -----------------------------
modelFit = model.fit(optimized=True)

# -----------------------------
# Generate forecasts
# -----------------------------
# January = 31 days * 24 hours = 744 hours
pred = modelFit.forecast(744)

# Convert predictions to numpy array (grader-friendly)
pred = np.array(pred)
```
